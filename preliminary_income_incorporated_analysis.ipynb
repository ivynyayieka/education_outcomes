```python
%%R -i df_edu_plus_income_for_analysis
# guide for multivariable regression
school_density_variable<-(df_edu_plus_income_for_analysis$population/df_edu_plus_income_for_analysis$number_of_schools)/df_edu_plus_income_for_analysis$land_area_km_sq

model <- lm(left_school_over_left_before_and_after_as_pct ~total_revenue_2021_22_plus + average_school_size+ pupil_textbook_ratio  + public_over_overall_enrolment_as_pct + school_density_variable, df_edu_plus_income_for_analysis)
summary(model)
```

```python
%%R 
# guide for multivariable regression
model <- lm(never_been_to_school_or_learning_institution_pct ~ log(average_school_size) + public_over_overall_enrolment_as_pct + with_national_grid_pct + lighting_Mains_Electricity + drinking_water_Piped_into_dwelling + Use_of_Internet_pct + Seeking_Work_No_Work_Available_over_Total_Total_pct, df_edu_plus_income_for_analysis)
summary(model)
```

```python
%%R -o df_edu_plus_income_for_analysis_residuals
#access to education
# guide for multivariable regression

df_edu_plus_income_for_analysis_residuals<- df_edu_plus_income_for_analysis %>% 
    mutate(residuals=resid(model),
          residuals_z=(residuals-mean(residuals))/sd(residuals))
```
